## The Technological Process
Before we dive into the data, let's briefly discuss the technological process of gold extraction.

### Flotation: 
The mined ore, called the rougher feed, undergoes flotation to obtain the rougher Au (gold) concentrate and rougher tails. The stability of this process depends on the volatile and non-optimal physicochemical state of the flotation pulp.

### Purification: 
The rougher concentrate is then subjected to two stages of purification to yield the final concentrate and new tails.

We'll be dealing with parameters from each of these stages, which will have been labeled as per their stage and parameter type.

### Recovery Calculation
An integral part of our project will be simulating the process of recovering gold from the ore. This will be done using a given formula that considers the share of gold in the concentrate right after flotation and after purification, in the feed before flotation and in the concentrate right after flotation, and in the rougher tails right after flotation and after purification.

### Evaluation Metric <br>
Our model's performance will be evaluated using the Symmetric Mean Absolute Percentage Error (sMAPE). sMAPE is similar to the Mean Absolute Error (MAE), but it is expressed in relative values instead of absolute ones, taking into account the scale of both the target and the prediction. We'll need to predict two values, the rougher concentrate recovery and the final concentrate recovery, which will be used to calculate the final sMAPE.

### Helper Function - Calculate Recovery

In order to calculate the recovery rate of gold from the gold ore, we will need a special function that considers the concentration of gold in the final concentrate, feed, and tails. Here is the `calculate_recovery` function: <br>
<br>


```python
def calculate_recovery(C, F, T):
    """
    Function to calculate the recovery of gold from gold ore.

    Parameters:
    C : float
        Concentration of gold in the final concentrate.
    F : float
        Concentration of gold in the feed/initial raw material.
    T : float
        Concentration of gold in the tails/rougher tails (residues).

    Returns:
    recovery : float
        Recovery rate of gold.
    """

    recovery = ((C * (F - T)) / (F * (C - T))) * 100 if F*(C-T) != 0 else 0
    return recovery
```

The above Function will return the recovery rate of gold. Note that it also includes a condition to prevent division by zero.

```python
import numpy as np

def calculate_sMAPE(y_true, y_pred):
    """
    Function to calculate the symmetric Mean Absolute Percentage Error (sMAPE).

    Parameters:
    y_true : array-like of shape (n_samples,)
        Ground truth (correct) target values.
    y_pred : array-like of shape (n_samples,)
        Estimated targets as returned by a classifier.

    Returns:
    sMAPE : float
        The sMAPE between y_true and y_pred
    """

    # Ensure the arrays are numpy arrays
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Calculate the sMAPE
    sMAPE = np.mean(np.abs(y_true - y_pred) / ((np.abs(y_true) + np.abs(y_pred)) / 2)) * 100

    return sMAPE
```

The above function takes in two arguments:

y_true - the actual values
y_pred - the predicted values
The function converts these inputs into numpy arrays (if they are not already), and calculates the sMAPE between the actual and predicted values. We will use this function to evaluate our model's performance later on.

In the following steps, we will start working with our data set. We will load it, clean it, and perform necessary transformations to prepare it for the modeling phase.

# Final sMAPE Calculation Function

To generate the final evaluation metric for our machine learning model, we need to compute the final sMAPE which is a combination of the sMAPEs for the rougher concentrate recovery and the final concentrate recovery. Here is the Python function `final_sMAPE` to calculate this:



```python
def final_sMAPE(y_true_rougher, y_pred_rougher, y_true_final, y_pred_final):
    """
    Function to calculate the final sMAPE for rougher and final concentrate recovery.

    Parameters:
    y_true_rougher : array-like of shape (n_samples,)
        Ground truth (correct) target values for rougher concentrate recovery.
    y_pred_rougher : array-like of shape (n_samples,)
        Estimated targets as returned by a classifier for rougher concentrate recovery.
    y_true_final : array-like of shape (n_samples,)
        Ground truth (correct) target values for final concentrate recovery.
    y_pred_final : array-like of shape (n_samples,)
        Estimated targets as returned by a classifier for final concentrate recovery.

    Returns:
    final_sMAPE : float
        The final sMAPE for rougher and final concentrate recovery.
    """

    # Calculate the sMAPE for rougher and final concentrate recovery
    sMAPE_rougher = calculate_sMAPE(y_true_rougher, y_pred_rougher)
    sMAPE_final = calculate_sMAPE(y_true_final, y_pred_final)

    # Calculate the final sMAPE
    final_sMAPE = 0.25 * sMAPE_rougher + 0.75 * sMAPE_final

    return final_sMAPE
```

The above function takes in four arguments:

- `y_true_rougher` - the actual values for the rougher concentrate recovery
- `y_pred_rougher` - the predicted values for the rougher concentrate recovery
- `y_true_final` - the actual values for the final concentrate recovery
- `y_pred_final` - the predicted values for the final concentrate recovery

The function calculates the sMAPEs for the rougher and final concentrate recovery using the `calculate_sMAPE` function we defined earlier. It then computes the final sMAPE as a weighted sum of the sMAPEs for the rougher concentrate recovery and the final concentrate recovery, with the latter carrying more weight.

This final sMAPE will serve as our main evaluation metric when assessing the performance of our model.

```python
import os
import pandas as pd


def download_and_save_data(urls, folder):
    """Download the source CSV files and save them in a specified folder."""

    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"Directory '{folder}' created.")

    for name, url in urls.items():
        file_path = f'{folder}/gold_recovery_{name}.csv'
        
        if not os.path.isfile(file_path):
            print(f"Downloading data from {url}...")
            data = pd.read_csv(url, parse_dates=['date'], index_col='date')
            data.to_csv(file_path)
            print(f"Data saved to {file_path}")
        else:
            print(f"File {file_path} already exists. No need to download.")


def load_data(urls, folder):
    """Load CSV files from a specified folder."""

    datasets = {}
    for name in urls.keys():
        print(f"Loading data from {folder}/gold_recovery_{name}.csv...")
        datasets[name] = pd.read_csv(f'{folder}/gold_recovery_{name}.csv', parse_dates=['date'], index_col='date')
    print("Data loading completed.")

    return datasets


def process_data():
    urls = {
        'train': 'https://practicum-content.s3.us-west-1.amazonaws.com/datasets/gold_recovery_train.csv',
        'test': 'https://practicum-content.s3.us-west-1.amazonaws.com/datasets/gold_recovery_test.csv',
        'full': 'https://practicum-content.s3.us-west-1.amazonaws.com/datasets/gold_recovery_full.csv'
    }

    folder = 'datasets'
    download_and_save_data(urls, folder)
    datasets = load_data(urls, folder)
    return


if __name__ == "__main__":
    datasets = process_data()

```

## Calculate recovery for the rougher.output.recovery feature and find the MAE


```python
# # Calculate recovery for the rougher.output.recovery feature and find the MAE

import numpy as np

# Define a function to calculate recovery
def calculate_recovery(c, f, t):
    epsilon = 1e-15  # Small constant to avoid division by zero
    recovery = (c * (f - t)) / ((f * (c - t)) + epsilon) * 100
    return recovery

# Calculate the recovery for the train_data
train_data['calculated_recovery'] = train_data.apply(lambda row: calculate_recovery(row['rougher.output.concentrate_au'], row['rougher.input.feed_au'], row['rougher.output.tail_au']), axis=1)

# Remove any potential NaN values after calculation
train_data = train_data.dropna(subset=['calculated_recovery', 'rougher.output.recovery'])

# Calculate the mean absolute error (MAE) between the calculated recovery and the given recovery values
mae = (abs(train_data['calculated_recovery'] - train_data['rougher.output.recovery'])).mean()
print("Mean Absolute Error (MAE) between calculated recovery and given recovery values in train_data:", mae)
```

To identify and eliminate anomalies, we can look for data points that are significantly lower than the first quartile (Q1) or significantly higher than the third quartile (Q3). You can calculate the interquartile range (IQR) and consider data points outside 1.5 * IQR below Q1 or above Q3 as potential outliers.


```python
# Set the IQR range
lower_percentile = 0.1
upper_percentile = 0.9

# Filter out rows with total concentrations outside the IQR range
filtered_train_data = train_data.copy()
filtered_test_data = test_data.copy()

for stage in stages:
    Q1 = train_data[f'{stage}_total_concentration'].quantile(lower_percentile)
    Q3 = train_data[f'{stage}_total_concentration'].quantile(upper_percentile)
    
    filtered_train_data = filtered_train_data[
        (filtered_train_data[f'{stage}_total_concentration'] >= Q1) & 
        (filtered_train_data[f'{stage}_total_concentration'] <= Q3)
    ]
    
    filtered_test_data = filtered_test_data[
        (filtered_test_data[f'{stage}_total_concentration'] >= Q1) & 
        (filtered_test_data[f'{stage}_total_concentration'] <= Q3)
    ]

# Check the shape of the filtered datasets
print("Filtered Training Data shape:", filtered_train_data.shape)
print("Filtered Test Data shape:", filtered_test_data.shape)
```

The above code block filters out the rows in the training and test datasets with total concentrations outside the specified interquartile range (IQR). It uses the lower and upper percentiles (0.1 and 0.9) to calculate the first quartile (Q1) and third quartile (Q3) of the total concentrations at each stage in the training data. Then, it retains only the rows with total concentrations within the IQR range for each stage in both datasets.

After filtering, the output shows the shapes of the filtered datasets. The filtered training data has 7,650 rows and 90 columns, while the filtered test data has 2,859 rows and 55 columns. This means that some rows with extreme total concentrations have been removed from the original datasets, potentially improving the model's performance by reducing the influence of outliers.

```python
def sMAPE(y_true, y_pred):
    epsilon = 1e-10  # you can adjust this value as needed
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    smape_value = 2 * np.mean(np.abs(y_true - y_pred) / (np.abs(y_true) + epsilon + np.abs(y_pred)))

    return smape_value

def final_sMAPE(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```
